```python
import pandas as pd
```

```python
df = pd.read_excel("data/flight_price.xlsx")
```

```python
df.shape
```

```python
df["Date_of_Journey"].str.split("/")
df["Date"] = df["Date_of_Journey"].str.split("/").str[0].astype("int")
df["Month"] = df["Date_of_Journey"].str.split("/").str[1].astype("int")
df["Year"] = df["Date_of_Journey"].str.split("/").str[2].astype("int")
```

```python
df.head()
```

```python
df.drop("Date_of_Journey", axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Arrival_Time"] = df["Arrival_Time"].str.split(" ").str[0]
```

```python
df["Arrival_hour"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_minute"] = df["Arrival_Time"].str.split(":").str[1]
```

```python
df.head()
```

```python
df.drop("Arrival_Time", inplace=True, axis=1)
```

```python
df["Arrival_hour"] = df["Arrival_hour"].astype("int")
df["Arrival_minute"] = df["Arrival_minute"].astype("int")
```

```python
df.info()
```

```python
df.head()
```

```python
df["Dep_Time_hour"] = df["Dep_Time"].str.split(":").str[0].astype("int")
df["Dep_Time_minute"] = df["Dep_Time"].str.split(":").str[1].astype("int")
df.drop("Dep_Time", inplace=True, axis=1)
```

```python
df.head()
```

```python
df.drop("Route",inplace=True,axis=1)
```

```python
df.head()
```

```python
df["Duration_hour"] = df["Duration"].str.split(" ").str[0].str.split("").str[1]
df["Duration_minute"] = df["Duration"].str.split(" ").str[1].str.split("").str[1]
```

```python
df.head()
```

```python
df["Duration_minute"].fillna(0, inplace=True)
```

```python
df.drop("Duration", inplace=True, axis=1)
```

```python
df.head()
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"].fillna("1 stop", inplace=True)
```

```python
df["Total_Stops"].isna().sum()
```

```python
df["Total_Stops"] = df["Total_Stops"].map({"non-stop" : 0, "1 stop" : 1, "2 stops" : 2, "3 stops" : 3, "4 stops" : 4})
```
