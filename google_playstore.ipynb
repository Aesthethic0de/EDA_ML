```python
df_copy = df.copy()
```

```python
df_copy.drop(df_copy.index[10472],inplace=True)
```

```python
df_copy["Reviews"].astype("int")
```

```python
df_copy["Reviews"]  = df_copy["Reviews"].astype("int")
```

```python
df_copy.info()
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"] = df_copy["Size"].str.replace("M", "000")
df_copy["Size"] = df_copy["Size"].str.replace("k", "")
```

```python
df_copy["Size"].unique()
```

```python
df_copy["Size"] = df_copy["Size"].replace("Varies with device", np.nan)
```

```python
df_copy["Installs"].unique()
```

```python
df_copy["Price"].unique()
```

```python
df_copy.head()
```

```python
chars_to_remmove = ["+", ",", "$"]
cols_to_clean = ["Installs", "Price"]
for item in chars_to_remmove:
    for cols in cols_to_clean:
        df_copy[cols] = df_copy[cols].str.replace(item, "")
```

```python
df_copy.head()
```

```python
df_copy["Installs"]=df_copy["Installs"].astype("int")
```

```python
df_copy["Price"]=df_copy["Price"].astype("float")
```

```python
df_copy["Price"].unique()
```

```python
df_copy.info()
```

```python
df_copy["Last Updated"] = pd.to_datetime(df_copy["Last Updated"])
df_copy["Day"] = df_copy["Last Updated"].dt.day
df_copy["Month"] = df_copy["Last Updated"].dt.month
df_copy["Year"] = df_copy["Last Updated"].dt.year
```

```python
df_copy.drop("Last Updated",axis=1, inplace=True)
```

```python
df_copy.head()
```

```python
df_copy["Content Rating"].unique()
```

```python
df_copy["Content Rating"].value_counts()
```

```python
one_hot_encoded = pd.get_dummies(data=df_copy, columns=["Content Rating"])
```

```python
one_hot_encoded.columns
```

```python
one_hot_encoded.info()
```

# EDA

## Dropped duplicated and kept only first

```python
one_hot_encoded[one_hot_encoded.duplicated("App")]
```

```python
one_hot_encoded.drop_duplicates(subset=["App"], keep="first",inplace=True)
```

Lets go and Explore numerical and categorical data

```python
numeric_features = [feature for feature in one_hot_encoded.columns if one_hot_encoded[feature].dtypes!="O"]
categorical_features = [feature for feature in one_hot_encoded.columns if one_hot_encoded[feature].dtypes=="O"]
```

## Visualization diagram

## Observation for numerical data

#### Ratings and year are right skewed

```python
# one_hot_encoded.columns = one_hot_encoded.columns.str.replace(' ', '')
one_hot_encoded
```

```python
plt.figure(figsize=(20,10))
plt.suptitle("Univariate analysis of Numerical features", fontsize=20, fontweight="bold", alpha=0.8, y=1.)
for i in range(len(numeric_features)):
    print(numeric_features[i])
    plt.subplot(5,3,i+1)
    sns.kdeplot(x=one_hot_encoded[numeric_features[i]],fill=True, color="r")
    plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))
plt.suptitle("Univariate analysis of Numerical features", fontsize=20, fontweight="bold", alpha=0.8, y=1.)
for i in range(len(numeric_features)):
    plt.subplot(5,3,i+1)
    sns.histplot(x=one_hot_encoded[numeric_features[i]],fill=True, color="r")
    plt.tight_layout()
```
